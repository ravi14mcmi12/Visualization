```python
import pandas as pd
import numpy as np
import seaborn as sb
from matplotlib import pyplot as plt
```

### 1. Plotting Univariate Distribution

```python
sb.distplot?
```

```python
df = sb.load_dataset('iris')
df.head()
```

```python
sb.distplot(df.petal_length, kde=False, color="Red")
```

```python
sb.distplot(df.petal_length, hist=False)
```

```python
sb.distplot(df.petal_length, kde=True, color = 'Red')
```

```python
sb.distplot(df.petal_length)
```

### 2. Plotting Bivariate Distribution

#### Scatter Plot

```python
# sb.jointplot(x = df.petal_length, y = df.petal_width)
sb.jointplot(x = 'petal_length', y = 'petal_width',  data = df)
```

### Hexbin Plot
 - An addition parameter called 'kind' and vlaue 'hex' plots the hexbin lot

```python
sb.jointplot(x = df.petal_length, y = df.petal_width, data=df, kind='hex', color='blue')
```

### Kerel Density Estimation

```python
sb.jointplot(x = df.petal_length, y = df.petal_width, data= df, kind='kde')
```
