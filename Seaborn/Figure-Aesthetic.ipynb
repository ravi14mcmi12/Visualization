```python
## Wrtie a sin wave function to plot different 
def sinplot(flip = 1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 5):
        plt.plot(x, np.sin(x+i*0.5)*(7-i)*flip)
#         plt.show()
```

```python
sinplot()
```

#### To chage the same plot to Seaborn Default

```python
sb.set()
sinplot()
```

### Basically, Seaborn splits the Matplotlob parameters into two groups
   - Plot Style
   - Plot scale

### Seaborn Figure Styles
 - Darkgrid
 - Whitegrid
 - Dark
 - White
 - Ticks

```python
figure_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']
# figure_styles = ['whitegrid']
for style in figure_styles:
    sb.set_style(style= style)
    sinplot()
    plt.show()
```
